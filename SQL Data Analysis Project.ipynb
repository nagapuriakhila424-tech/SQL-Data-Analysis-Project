 STEP 1: IMPORT LIBRARIES


```python
import pandas as pd
import sqlite3
```

 STEP 2: LOAD DATASET



```python
print("LOAD DATASET")
df = pd.read_csv(r"Dataset for SQL Data Analysis.csv")
print("Dataset Shape:", df.shape)
print("\nDATA PREVIEW:")
print(df.head())
print("\nColumns:")
print(df.columns.tolist())
```

 STEP 3: DATA QUALITY CHECK



```python
print("DATA QUALITY CHECK")
print("\nMissing Values:")
print(df.isnull().sum())
print("\nDuplicate Rows:")
print(df.duplicated().sum())
# STEP 3A: DATA CLEANING
print("DATA CLEANING")
df = df.drop_duplicates()
df = df.dropna()
df.columns = df.columns.str.strip()
df.to_csv("Cleaned_Dataset.csv", index=False)
print("Cleaned Dataset Shape:", df.shape)
print("Cleaned Dataset Saved Successfully!")
```

 STEP 4: CREATE SQLITE DATABASE


```python
print("CREATE SQLITE DATABASE")
conn = sqlite3.connect("data_analytics.db")
```

STEP 5: CREATE SQL TABLE


```python
print("CREATE SQL TABLE")
df.to_sql("orders", conn, if_exists="replace", index=False)
print("Table 'orders' created successfully!")
```

 STEP 6: SELECT QUERY


```python
print("SELECT QUERY")
query = """
SELECT *
FROM orders
LIMIT 5
"""
print(pd.read_sql(query, conn))
```

 STEP 7: WHERE CLAUSE


```python
print("WHERE CLAUSE")
query = """
SELECT *
FROM orders
WHERE TotalPrice > 2000
"""
print(pd.read_sql(query, conn))
```

 STEP 8: ORDER BY CLAUSE


```python
print("ORDER BY CLAUSE")
query = """
SELECT *
FROM orders
ORDER BY TotalPrice DESC
LIMIT 5
"""
print(pd.read_sql(query, conn))
```

 STEP 9: GROUP BY CLAUSE


```python
print("GROUP BY CLAUSE")
query = """
SELECT Product,
       COUNT(*) AS Order_Count
FROM orders
GROUP BY Product
"""
print(pd.read_sql(query, conn))
```

 STEP 10: COUNT() AGGREGATION


```python
print("COUNT() AGGREGATION")
query = """
SELECT COUNT(*) AS Total_Orders
FROM orders
"""
print(pd.read_sql(query, conn))
```

 STEP 11: SUM() AGGREGATION


```python
print("SUM() AGGREGATION")
query = """
SELECT SUM(TotalPrice) AS Total_Revenue
FROM orders
"""
print(pd.read_sql(query, conn))
```

 STEP 12: AVG() AGGREGATION


```python
print("AVG() AGGREGATION")
query = """
SELECT AVG(TotalPrice) AS Average_Order_Value
FROM orders
"""
print(pd.read_sql(query, conn))
```

STEP 13: REVENUE BY PRODUCT


```python
print("REVENUE BY PRODUCTS")
query = """
SELECT Product,
       SUM(TotalPrice) AS Revenue 
FROM orders
GROUP BY Product
ORDER BY Revenue DESC
"""
print(pd.read_sql(query, conn))
```

 STEP 14: ORDERS BY PAYMENT METHOD


```python
print("ORDERS BY PAYMENT METHOD")
query = """
SELECT PaymentMethod,
       COUNT(*) AS Orders
FROM orders
GROUP BY PaymentMethod
"""
print(pd.read_sql(query, conn))
```

STEP 15: ORDERS BY STATUS


```python
print("ORDERS BY STATUS")
query = """
SELECT OrderStatus,
       COUNT(*) AS Total_Orders
FROM orders
GROUP BY OrderStatus
"""
print(pd.read_sql(query, conn))
```

STEP 15A: WHERE + GROUP BY + ORDER BY COMBINED


```python
print("ADVANCED SQL QUERY")

query = """
SELECT Product,
       COUNT(*) AS Total_Orders,
       SUM(TotalPrice) AS Revenue
FROM orders
WHERE TotalPrice > 2000
GROUP BY Product
ORDER BY Revenue DESC
"""

print(pd.read_sql(query, conn))
```

 STEP 16: TOP 5 HIGHEST VALUE ORDERS


```python
print("TOP 5 HIGHEST VALUE ORDERS")
query = """
SELECT OrderID,
       Product,
       TotalPrice
FROM orders
ORDER BY TotalPrice DESC
LIMIT 5
"""
print(pd.read_sql(query, conn))
```

STEP 17: BUSINESS INSIGHTS


```python
print("BUSINESS INSIGHTS")

query = """
SELECT Product,
       SUM(TotalPrice) AS Revenue
FROM orders
GROUP BY Product
ORDER BY Revenue DESC
LIMIT 1
"""

top_product = pd.read_sql(query, conn)

print("\nHighest Revenue Generating Product:")
print(top_product)

print("\nINSIGHT:")
print(f"The product '{top_product.iloc[0]['Product']}' generated the highest revenue of ₹{top_product.iloc[0]['Revenue']:.2f}.")
```
